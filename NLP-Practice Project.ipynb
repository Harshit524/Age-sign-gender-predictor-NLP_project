```python
import re
df.text = df.text.apply(lambda x: re.sub('[^A-Za-z]+', ' ', x))   # for selecting only the alphabets

# Convert text to lowercase
df.text = df.text.apply(lambda x: x.lower())

# Strip unwanted spaces
df.text = df.text.apply(lambda x: x.strip())

#took the help from the solutions 
```

```python
import nltk
```

```python
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
```

```python
stopwords
```

```python
stopwords= set(stopwords.words('english'))          #took the help from solution
```

```python
len(stopwords)        #there are 179 stopwords
```

```python
df.text = df.text.apply(lambda x: ' '.join([word for word in x.split() if word not in stopwords]))  #taken help from solution 
```

```python
df
```

```python
df.drop('date', axis= 1, inplace= True)
```

```python
df.shape
```

```python
df['labels']= df.apply(lambda row: [row['gender'], str(row['age']), row['topic'], row['sign']], axis=1)
```

```python
df.drop(['gender', 'age', 'sign', 'topic'], axis= 1, inplace= True)  #removing the individual columns after forming a cosolidated column 
```

```python
df.head()
```

```python
#splitting the test and train data set 
X= df['text']
y= df['labels']
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(X, y, random_state= 1, test_size= 0.3)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
toNumeric=  CountVectorizer(binary= True, ngram_range=(1, 2))
```

```python
toNumeric.fit(X_train)
```

```python
dtm_train= toNumeric.transform(X_train)
dtm_test= toNumeric.transform(X_test)
```

```python
label_counts = dict()

for labels in df.labels.values:
    for label in labels:
        if label in label_counts:
            label_counts[label] += 1
        else:
            label_counts[label] = 1
```

```python
label_counts
```

```python
from sklearn.preprocessing import MultiLabelBinarizer
MLB= MultiLabelBinarizer(classes= sorted(label_counts.keys()))

y_train= MLB.fit_transform(y_train)
y_test= MLB.transform(y_test)
```

```python
from sklearn.multiclass import OneVsRestClassifier
from sklearn.linear_model import LogisticRegression

clf = LogisticRegression(solver='lbfgs')
clf = OneVsRestClassifier(clf)
```

```python
logreg_model= clf.fit(dtm_train, y_train)
```

```python
y_pred= logreg_model.predict(dtm_test)
```

```python
y_pred
```

```python
pred_inverse= MLB.inverse_transform(y_pred)
y_test_inversed = MLB.inverse_transform(y_test)
```

```python
from sklearn.metrics import accuracy_score, f1_score, average_precision_score, recall_score
```

```python
accuracy_score(y_test, y_pred)
```

```python
f1_score(y_test, y_pred, average= 'micro' )
```

```python
average_precision_score(y_test, y_pred, average= 'micro')
```

```python
recall_score(y_test, y_pred, average= 'micro')
```
